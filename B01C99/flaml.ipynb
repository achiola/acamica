```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('DS_Proyecto_01_Datos_Properati.csv')
```

```python
X = df.drop(['price'],axis=1)
y = df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=1975)
```

```python
from flaml import AutoML
```

```python
automl = AutoML()
automl.fit(X_train, y_train, task="regression", time_budget=3000)
```

```python
print(automl.model)
```

```python
predicciones = automl.predict(X_test)
```

```python
X_test['real'] = y_test
X_test['predicho'] = predicciones
```
